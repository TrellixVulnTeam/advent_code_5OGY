All numbers in the elves' list are in feet. How many total square feet of wrapping paper should they order?

```python
class Present:
    """A simple model of a present"""
    def __init__(self, l, w, h):
        l = int(l)
        w = int(w)
        h = int(h)
        self.side_1 = l * w
        self.ribbon_1 = 2*l + 2*w
        self.side_2 = w * h
        self.ribbon_2 = 2*w + 2*h
        self.side_3 = h * l
        self.ribbon_3 = 2*h + 2*l
        self.bow = l * w * h
        self.sides = [self.side_1, self.side_2, self.side_3]
        self.ribbon = [self.ribbon_1, self.ribbon_2, self.ribbon_3]
        self.slack = min(self.sides)
        self.ribbon_bow_and_perimeter = min(self.ribbon) + self.bow
        self.sides_total = 2 * (sum(self.sides))
        self.sides_total += self.slack
```

```python
with open(file_dir) as f:
    presents = f.read()
    present_list = [dimensions.split('x') for dimensions in presents.split()]
    paper_area = 0
    ribbon_total = 0
    for measures in present_list:
        this_present = Present(
            l=int(measures[0]), w=int(measures[1]), h=int(measures[2])
            )
        paper_area += this_present.sides_total
        ribbon_total += this_present.ribbon_bow_and_perimeter
    print(paper_area)
    print(ribbon_total)
```

A present with dimensions 1x1x10 requires 2*1 + 2*10 + 2*10 = 42 square feet of wrapping paper plus 1 square foot of slack, for a total of 43 square feet.

All numbers in the elves' list are in feet. How many total square feet of wrapping paper should they order?

```python
file_dir = %pwd
file_dir += '/Day_2_Part_1.txt'
presents_list = open(file_dir).read()
presents = presents_list.split()
presents = [dimensions.split('x') for dimensions in presents]
```

```python
def get_paper(dimensions=[]):
    d = list(map(int, dimensions))
    sides = [d[0]*d[1], d[1]*d[2], d[0]*d[2]]
    slack = min(sides)
    total = 2 * (sides[0] + sides[1] + sides[2]) + slack
    return total

present_paper = sum([get_paper(present) for present in presents])
print(present_paper)
```

He begins by delivering a present to the house at his starting location, and then an elf at the North Pole calls him via radio and tells him where to move next.
```python
file_dir = %pwd
file_dir += '/Day_3.txt'
directions_list = open(file_dir).read().strip()
directions = []
directions[::] = directions_list

class Santa:
    """A model of santa"""
    def __init__(self):
        self.x = 0
        self.y = 0
        self.visited = []
    
    def location(self):
        return (self.x, self.y)
    
    def go_north(self):
        self.y = self.y + 1
    
    def go_south(self):
        self.y  = self.y - 1
    
    def go_east(self):
        self.x += 1
    
    def go_west(self):
        self.x = self.x -1
    
    def take_step(self, step):
        if step == '>':
            self.go_east()
        elif step == "<":
            self.go_west()
        elif step == 'v':
            self.go_south()
        else:
            self.go_north()
    
    def mark_visited(self):
        now = self.location()
        self.visited.append(now)
        
    def follow_directions(self, directions=[], start='', skip=''):
        for direction in directions[start::skip]:
            self.take_step(direction)
            self.mark_visited()
        
santa = Santa()
robot = Santa()
santa.mark_visited()
robot.mark_visited()
santa.follow_directions(directions=directions, start=0, skip=2)
robot.follow_directions(directions=directions, start=1, skip=2)
houses = santa.visited + robot.visited
unique_houses = set(houses)
len(unique_houses)
```

--- Day 5: Doesn't He Have Intern-Elves For This? ---

Santa needs help figuring out which strings in his text file are naughty or nice.

A nice string is one with all of the following properties:

    It contains at least three vowels (aeiou only), like aei, xazegov, or aeiouaeiouaeiou.
    It contains at least one letter that appears twice in a row, like xx, abcdde (dd), or aabbccdd (aa, bb, cc, or dd).
    It does not contain the strings ab, cd, pq, or xy, even if they are part of one of the other requirements.

For example:

    ugknbfddgicrmopn is nice because it has at least three vowels (u...i...o...), a double letter (...dd...), and none of the disallowed substrings.
    aaa is nice because it has at least three vowels and a double letter, even though the letters used by different rules overlap.
    jchzalrnumimnmhp is naughty because it has no double letter.
    haegwjzuvuyypxyu is naughty because it contains the string xy.
    dvszwmarrgswjxmb is naughty because it contains only one vowel.

How many strings are nice?


```python
file_dir = %pwd
file_dir += '/Day_5.txt'
strings = open(file_dir).read().splitlines()

def check_vowels_nice(string):
    vowel_count = 0
    vowels = ['a','e','i','o','u']
    characters = []
    characters[::] = string
    for character in characters:
        if character in vowels:
            vowel_count += 1
    if vowel_count >= 3:
        return True
    else:
        return False
    
def check_doubles_nice(string):
    """Check if a character is doubled."""
    characters = []
    characters[::] = string
    for character in characters:
        if f"{character}{character}" in string:
            return True
        else:
            continue
    return False

def check_no_naughties(string):
    """Check if a string is not naughty."""
    naughty_strings = ['ab', 'cd', 'pq', 'xy']
    for naughty in naughty_strings:
        if naughty in string:
            return False
    else:
        return True

def check_string(string):
    first_check = check_vowels_nice(string)
    second_check = check_doubles_nice(string)
    third_check = check_no_naughties(string)
    checks = [first_check, second_check, third_check]
    if False not in checks:
        return True
    
def check_all_strings(string_list):
    good_strings = []
    for string in string_list:
        response = check_string(string)
        if response == True:
            good_strings.append(string)
    return len(good_strings)

task_1 = check_all_strings(strings)
print(task_1)
```

Because your neighbors keep defeating you in the holiday house decorating contest year after year, you've decided to deploy one million lights in a 1000x1000 grid.

The instructions include whether to turn on, turn off, or toggle various inclusive ranges given as coordinate pairs. Each coordinate pair represents opposite corners of a rectangle, inclusive; a coordinate pair like 0,0 through 2,2 therefore refers to 9 lights in a 3x3 square. The lights all start turned off.

To defeat your neighbors this year, all you have to do is set up your lights by doing the instructions Santa sent you in order.

For example:

    turn on 0,0 through 999,999 would turn on (or leave on) every light.

After following the instructions, how many lights are lit?

```python
file_dir = %pwd
file_dir += '/Day_6.txt'
directions = open(file_dir).read().splitlines()
direction_length = len(directions)

class Light:
    """A model of a single light."""
    def __init__(self, x, y, lit=0):
        self.x = int(x)
        self.y = int(y)
        self.lit = 0
        
    def toggle_on(self):
        self.lit += 1
        
    def toggle_off(self):
        if self.lit > 0:
            self.lit -= 1
        
    def toggle(self):
        self.lit += 2

def create_lights():
    """Make all the light objects and add to a list."""
    lights = []
    for x in range(0, 1000):
        for y in range(0, 1000):
            light = Light(x=x, y=y, lit=0)
            lights.append(light)
    return lights
    
def light_instructions(string, lights):
    """Take a string from the directions list, parse the numbers, and flip lights."""
    directions = string.split(' ')
    length = len(directions)
    first_value = directions[length-3]
    first_list = first_value.split(',')
    first_x = int(first_list[0])
    first_y = int(first_list[1])
    second_value = directions[length-1]
    second_list = second_value.split(',')
    second_x = int(second_list[0]) + 1
    second_y = int(second_list[1]) + 1
    if directions[1] == 'on':
        for light in lights:
            if light.x in range(first_x, second_x):
                if light.y in range(first_y, second_y):
                    light.toggle_on()
    elif directions[1] == 'off':
        for light in lights:
            if light.x in range(first_x, second_x):
                    if light.y in range(first_y, second_y):
                        light.toggle_off()
    else:
        for light in lights:
            if light.x in range(first_x, second_x):
                if light.y in range(first_y, second_y):
                    light.toggle()
            
lights = create_lights()
string_counter = 0
for string in directions:
    print(string)
    light_instructions(string, lights)
    string_counter += 1
    print(f"\tDirection {string_counter} of {direction_length} complete.")
    total_brightness = 0
    for light in lights:
        total_brightness += light.lit
    print(f"\tCurrent brightness: {total_brightness}")
end_brightness = 0
for light in lights:
    end_brightness += light.lit
print(f"Final Brightness: {end_brightness}")

        
        
        
        
```
